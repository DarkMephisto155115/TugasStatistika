```python
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset dari file CSV
file_path = 'Auto Sales data.csv'
dataset = pd.read_csv(file_path)

# Pilih kolom yang akan diuji
column_names = ['ORDERNUMBER', 'QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER', 'SALES', 'DAYS_SINCE_LASTORDER', 'MSRP']

# Pisahkan dataset menjadi dua kelompok dengan menggunakan indeks
n_samples = len(dataset)
group_A = dataset.loc[:n_samples // 2, column_names]
group_B = dataset.loc[n_samples // 2:, column_names]

# Lakukan uji t untuk setiap kolom
for column in column_names:
    t_statistic, p_value = ttest_ind(group_A[column], group_B[column])

    # Hitung nilai uji statistik dan nilai p-value
    print(f'T-Statistic for {column}: {t_statistic}')
    print(f'P-Value for {column}: {p_value}')

    # Output hasil uji hipotesis
    alpha = 0.05
    if p_value < alpha:
        print(f'Reject H0 for {column}: Terdapat cukup bukti untuk menyatakan bahwa rata-rata tidak sama.')
        print(f'Accept H1 for {column}: Terdapat perbedaan yang signifikan antara kelompok A dan B.')
    else:
        print(f'Fail to reject H0 for {column}: Tidak cukup bukti untuk menyatakan bahwa rata-rata sama.')
        print(f'Fail to accept H1 for {column}: Tidak cukup bukti untuk menyatakan perbedaan yang signifikan antara kelompok A dan B.')

    # Membuat dan menampilkan kurva distribusi probabilitas
    plt.figure(figsize=(10, 6))

    sns.kdeplot(group_A[column], label='Group A', fill=True)
    sns.kdeplot(group_B[column], label='Group B', fill=True)
    plt.axvline(t_statistic, color='red', linestyle='dashed', linewidth=2, label='T-Statistic')
    
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```
